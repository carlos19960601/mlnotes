```python
import os
import pandas as pd

HOUSING_PATH = os.path.join("datasets", "global temperature")

def load_temperature_data(housing_path=HOUSING_PATH):
    csv_path = os.path.join(housing_path, "GlobalLandTemperaturesByMajorCity.csv")
    return pd.read_csv(csv_path)
```
