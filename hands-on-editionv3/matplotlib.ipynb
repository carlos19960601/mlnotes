# 画第一张图
导入 `matplotlib` 包

```python
import matplotlib
```

现在画第一张图

```python
import matplotlib.pyplot as plt

plt.plot([1,2,4,9,5,3])
plt.show()
```

**注意**：
> Matplotlib输出图像的时候可以使用多个库比如Tk, wxPython等，当在命令行中运行python时，必须指定库，比如matplotlib.use("TKAgg")。在Jupyter notebook中，`import matplotlib.pyplot`时会自动注册jupyter为图像输出

`polt`函数如果传入一个数组，会作为Y轴的数值。也可以传入2个数组，分为时X轴和Y轴数据

```python
plt.plot([-3, -2, 5, 0], [1, 6, 4, 3])
plt.show()
```

现在来画一个函数图像

```python
import numpy as np

x = np.linspace(-2,2,500)
y =x**2

plt.plot(x, y)
plt.show()
```

添加一点标题，x,y轴的标签，画个网格

```python
plt.plot(x,y)
plt.title("Square Function")
plt.xlabel("x")
plt.ylabel("y = x**2")
plt.grid(True)
plt.show()
```

# 线条样式和颜色


默认情况下，matplotlib会在连续的2个点之间画一条线

```python
plt.plot([0, 100, 100, 0, 0, 100, 50, 0, 100], 
          [0, 0, 100, 100, 0, 100, 130, 100, 0])
plt.axis([-10, 110, -10, 140]) # 指定x,y轴的范围
plt.show()
```

可以传第3个参数改变线条样式和颜色，比如"g--" 代表 "绿色点线"

```python
plt.plot([0, 100, 100, 0, 0, 100, 50, 0, 100], 
          [0, 0, 100, 100, 0, 100, 130, 100, 0], 
         "g--")
plt.axis([-10, 110, -10, 140])
plt.show()
```

同时画多种, 按照这种格式传 x1, y1, [style1], x2, y2, [style2], ...

```python
plt.plot([0, 100, 100, 0, 0], [0, 0, 100, 100, 0], "r-",
         [0, 100, 50, 0, 100], [0, 100, 130, 100, 0], "g--")
plt.axis([-10, 110, -10, 140])
plt.show()
```

或者多次调用`plot`方法

```python
plt.plot([0, 100, 100, 0, 0], [0, 0, 100, 100, 0], "r-")
plt.plot([0, 100, 50, 0, 100], [0, 100, 130, 100, 0], "g--")         
plt.axis([-10, 110, -10, 140])
plt.show()
```

你也可以画其他样子的，可以参考matplotlib文档，改变颜色和样式

```python
x = np.linspace(-1.4, 1.4, 30)
plt.plot(x, x, 'g--', x, x**2, 'r:', x, x**3, 'b^')
plt.show()
```

`plot`会返回`Line2D`对象，可以改变线的不透明度，宽度等等

```python
x = np.linspace(-1.4, 1.4, 30)
line1, = plt.plot(x, x, 'g--') 
line2, = plt.plot(x, x**2, 'r:')
line3, = plt.plot(x, x**3, 'b^')
line1.set_linewidth(3.0)
line1.set_dash_capstyle("round")
line3.set_alpha(0.2)
plt.show()
```

# Subplots
matplotlib图可以包含多个subplots，只需要调用`subplot`方法，subplots按照网格的形式展示，从上到下，从左到右，编号为1，2，3,....,n

```python
x = np.linspace(-1.4, 1.4, 30)
plt.subplot(2, 2, 1)  # 2 rows, 2 columns, 1st subplot = top left
plt.plot(x, x)
plt.subplot(2, 2, 2)  # 2 rows, 2 columns, 2nd subplot = top right
plt.plot(x, x**2)
plt.subplot(2, 2, 3)  # 2 rows, 2 columns, 3rd subplot = bottom left
plt.plot(x, x**3)
plt.subplot(2, 2, 4)  # 2 rows, 2 columns, 4th subplot = bottom right
plt.plot(x, x**4)
plt.show()
```

subplot(223) 是subplot(2,2,3)的简写

像这样创建跨span的图

```python
plt.subplot(2, 2, 1)  # 2 rows, 2 columns, 1st subplot = top left
plt.plot(x, x)
plt.subplot(2, 2, 2)  # 2 rows, 2 columns, 2nd subplot = top right
plt.plot(x, x**2)
# plt.subplot(2, 2, (3,4))
plt.subplot(2, 1, 2)  # 2 rows, *1* column, 2nd subplot = bottom
plt.plot(x, x**3)
plt.show()
```

subplot2grid创建更复杂的布局

```python
plt.subplot2grid((3,3), (0, 0), rowspan=2, colspan=2)
plt.plot(x, x**2)
plt.subplot2grid((3,3), (0, 2))
plt.plot(x, x**3)
plt.subplot2grid((3,3), (1, 2), rowspan=2)
plt.plot(x, x**4)
plt.subplot2grid((3,3), (2, 0), colspan=2)
plt.plot(x, x**5)
plt.show()
```

# 多图

matplotlib自动调用figure(1)，你可以自主切换figure

```python
x = np.linspace(-1.4, 1.4, 30)

plt.figure(1)
plt.subplot(211)
plt.plot(x, x**2)
plt.title("Square and Cube")
plt.subplot(212)
plt.plot(x, x**3)

plt.figure(2, figsize=(10, 5))
plt.subplot(121)
plt.plot(x, x**4)
plt.title("y = x**4")
plt.subplot(122)
plt.plot(x, x**5)
plt.title("y = x**5")


plt.figure(1)      # back to figure 1, current subplot is 212 (bottom)
plt.plot(x, -x**3, "r:")

plt.show()
```

```python
x = np.linspace(-2, 2, 200)
fig1, (ax_top, ax_bottom) = plt.subplots(2, 1, sharex=True)
fig1.set_size_inches(10,5)
line1, line2 = ax_top.plot(x, np.sin(3*x**2), "r-", x, np.cos(5*x**2), "b-")
line3, = ax_bottom.plot(x, np.sin(3*x), "r-")
ax_top.grid(True)

fig2, ax = plt.subplots(1, 1)
ax.plot(x, x**2)
plt.show()
```

```python
# Drawing text
```

```python
x = np.linspace(-1.5, 1.5, 30)
px = 0.8
py = px**2

plt.plot(x, x**2, "b-", px, py, "ro")

plt.text(0, 1.5, "Square function\n$y = x^2$", fontsize=20, color='blue', ha="center")
plt.text(px - 0.08, py, "Beautiful point", ha="right", weight="heavy")
plt.text(px + 0.05, py - 0.4, "x = %0.2f\ny = %0.2f"%(px, py), rotation=-30,
         color='gray')

plt.show()
```

```python
plt.plot(x, x**2, px, py, "ro")
plt.annotate("Beautiful point", xy=(px, py), xytext=(px-1.3,py+0.5),
                           color="green", weight="heavy", fontsize=14,
                           arrowprops={"facecolor": "lightgreen"})
plt.show()
```

```python
plt.plot(x, x**2, px, py, "ro")

bbox_props = dict(boxstyle="rarrow,pad=0.3", ec="b", lw=2, fc="lightblue")
plt.text(px-0.2, py, "Beautiful point", bbox=bbox_props, ha="right")

bbox_props = dict(boxstyle="round4,pad=1,rounding_size=0.2", ec="black",
                  fc="#EEEEFF", lw=5)
plt.text(0, 1.5, "Square function\n$y = x^2$", fontsize=20, color='black',
         ha="center", bbox=bbox_props)

plt.show()
```
