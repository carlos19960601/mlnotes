```python
import torch
torch._dynamo.config.cache_size_limit = 64
torch._dynamo.config.suppress_errors = True
torch.set_float32_matmul_precision('high')

import ChatTTS
from IPython.display import Audio

chat = ChatTTS.Chat()
chat.load_models()

texts = ["王佩佩，你是不是？"]
        
wavs = chat.infer(texts)
Audio(wavs[0], rate=24_000, autoplay=True)
```

```python
texts = ["我是曾强, [uv_break] [uv_break] 王佩佩[uv_break][uv_break] 我选你[laugh]"]

rand_spk = chat.sample_random_speaker(seed=2222)

params_infer_code = {'prompt':'[speed_2]', 'temperature':.3, 'spk_emb' : rand_spk}
params_refine_text = {'prompt':'[oral_8][laugh_0][break_6]'}

wavs = chat.infer(texts, params_infer_code=params_infer_code, params_refine_text=params_refine_text)
Audio(wavs[0], rate=24_000, autoplay=True)
```
